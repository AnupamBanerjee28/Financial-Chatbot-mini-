```python
# Import Pandas:
import pandas as pd
```

```python
# Load the Data:
df = pd.read_csv("C:/Users/baner/OneDrive/Documents/10kfillings.csv")
df
```

```python
# Remove Commas and Dollar Signs
df['total_revenue'] = df['Total Revenue'].replace({'\$': '', ',': ''}, regex=True).astype(int)
df['net_income'] = df['Net Income'].replace({'\$': '', ',': ''}, regex=True).astype(int)
df['total_assets'] = df['Total Assets'].replace({'\$': '', ',': ''}, regex=True).astype(int)
df['total_liabilities'] = df['Total Liabilities'].replace({'\$': '', ',': ''}, regex=True).astype(int)
df['cashflow'] = df['Cash Flow from Operating Activities'].replace({'\$': '', ',': ''}, regex=True).astype(int)
```

```python
# Calculate year-over-year changes for each financial metric:
df['Revenue Growth (%)'] = df.groupby('Company')['total_revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby('Company')['net_income'].pct_change() * 100
df['Total Assets Growth (%)'] = df.groupby('Company')['total_assets'].pct_change() * 100
df['Total Liabilities Growth (%)'] = df.groupby('Company')['total_liabilities'].pct_change() * 100
df['Cash Flow Growth (%)'] = df.groupby('Company')['cashflow'].pct_change() * 100
```

```python
# Summary statistics
summary = df.groupby('Company').agg({
    'total_revenue': 'mean',
    'net_income': 'mean',
    'total_assets': 'mean',
    'total_liabilities': 'mean',
    'cashflow': 'mean'
})
```
